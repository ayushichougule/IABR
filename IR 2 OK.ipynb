```python
!pip install pgmpy

import pandas as pd
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
from sklearn.preprocessing import LabelEncoder
import numpy as np

df=pd.read_csv("heart.csv")
df.head()


print(f"Missing Values: {df.isnull().sum()}")

df['age']=pd.cut(df['age'],bins=3,labels=['Young','Middle','Old'])
df['chol']=pd.cut(df['chol'],bins=3,labels=['Low','Medium','High'])
df.head()

model=BayesianNetwork([('age','target'),
                        ('chol','target'),
                        ('cp','target'),
                        ('target','thalach')])

model.edges()

model.fit(df,estimator=MaximumLikelihoodEstimator)

infer=VariableElimination(model)

result=infer.query(variables=['target'],evidence={'age':'Old','chol':'High'})
print(result)

result=infer.query(variables=['target'],evidence={'age':'Middle','chol':'Medium','cp':2})
print(result)
```
